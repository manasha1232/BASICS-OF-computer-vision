```python
import cv2
import os
# Path to video
video_path = os.path.join('.', 'Recording 2025-11-20 184205.mp4')
# Open video
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    print("Error: Could not open video")
    exit()
while True:
    ret, frame = cap.read()

    if not ret:
        break

    cv2.imshow("Video", frame)

    # Press 'q' to quit
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
# 0 = default webcam
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Error: Could not open webcam")
    exit()
while True:
    ret, frame = cap.read()
    if not ret:
        break
    cv2.imshow("Webcam", frame)
    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
##colorspaces in OpenCV
import cv2
import os
image_path = os.path.join('.','Screenshot 2025-11-20 173932.png')
img = cv2.imread(image_path)
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_bw=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

cv2.imshow("Image",img_bw)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#blurring an image
import cv2
import os
image_path = os.path.join('.','Screenshot 2025-11-20 173932.png')
img = cv2.imread(image_path)
k_size=7
img_blur=cv2.blur(img,(k_size,k_size))
img_gausblur=cv2.GaussianBlur(img,(k_size,k_size),3)
img_gausblur=cv2.medianBlur(img,k_size)


cv2.imshow("Image",img_gausblur)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
